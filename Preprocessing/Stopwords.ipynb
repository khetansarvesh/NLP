# Removing Stopwords

```python
import nltk
from nltk.tokenize import word_tokenize
# for word tokenization you need the 'punkt' package from NLTK data
nltk.download('punkt')
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

#instead of using these stopwords you can create a file of your own stopwords and upload that and use those as stopwords
# stop_words = pd.read_csv('stopwords.txt').iloc[:,0].values
```

```python
words_no_stopwords = [word for word in nltk.word_tokenize("Hello everyone, You have to build a very working pipeline.") if word not in stop_words]
print(words_no_stopwords)
```
