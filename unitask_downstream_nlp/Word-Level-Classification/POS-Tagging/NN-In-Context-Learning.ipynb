# **Using Spacy Library**
Spacy is not just restricted to english language, it supports a bunch of languages and you can find those in the spacy documentation

```python
# Spacy is by default installed in the root kernel of google colab hence you just need to import it, but if you are running spacy on you local computer then you can install spacy using following command -- pip install -U spacy
import spacy
```

```python
# importing the pretrained model - 'en_core_web_sm', change this name if you want to import some other pretrained model
model = spacy.load("en_core_web_sm")
```

```python
# passing the test text for which i want a POS tagging
doc = model("Apple is looking at buying U.K. startup for $1 billion")

# the ans gets stored in doc
```

```python
for token in doc:
  print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,token.shape_, token.is_alpha, token.is_stop)

# token.text -> actual token
# tokekn.lemma -> lemmetization of that token
# token.pos -> POS tag of that token
# token.tag ->
# token.dep -> dependency of that token
# token.shape -> tells if token is lower or upper case
# token.is_stop -> tells if token is a stopword or not
```
