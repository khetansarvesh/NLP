#**Using Spacy Library**
Spacy is a Library containing pretrained models, trained on standard datasets  

```python
# Spacy is by default installed in the root kernel of google colab hence you just need to import it but if you are running spacy on you local computer then you can install spacy using following command pip install -U spacy
import spacy
```

```python
# importing the pretrained model - 'en_core_web_sm', change this name if you want to import some other pretrained model
model = spacy.load("en_core_web_sm")
```

```python
# passing the test text for which i want a POS tagging
doc = model("Apple is looking at buying U.K. startup for $1 billion")

# the ans gets stored in doc
```

```python
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

#**Using NLTK Library**

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
from nltk.tag import pos_tag
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.chunk import ne_chunk
#FACILITY, GPE, GSP, LOCATION, ORGANIZATION, PERSON
#GPE is Geo-Political Entity
#GSP is Geo-Socio-Political group
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
entity=ne_chunk(pos_tag(word_tokenize("Apple is looking at buying U.K. startup for $1 billion")))
print(entity.pos())
```

```python
entity = ne_chunk([('I', 'PRP'), ('am', 'VBP'), ('Sam', 'NNP')])
print(entity.pos())
print('---------------')
print([pos[0][0] for pos in entity.pos() if pos[1] == 'PERSON'])
```
